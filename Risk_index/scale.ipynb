# 데이터 불러오기

```python
import pandas as pd
```

```python
data = pd.read_csv('/content/dataset_사용 - before_scale.csv')
```

```python
data.head()
```

# 데이터 전처리

```python
data = data.drop(['district'], axis=1) # 자치구 컬럼 삭제
```

```python
data = data.drop(['distance'], axis=1) # 거리두기 컬럼 삭제
```

```python
data.head() 
```

# min-max scaler

```python
from sklearn.preprocessing import MinMaxScaler
# 객체 생성
minmax_scaler = MinMaxScaler()
# 분포 저장
minmax_scaler.fit(data)
# 스케일링
df_minmax_scaled = minmax_scaler.transform(data)
```

```python
df_minmax_scaled = pd.DataFrame(df_minmax_scaled)
df_minmax_scaled
```

```python
df_minmax_scaled.min() # 확인
```

```python
df_minmax_scaled.max() # 확인
```

```python
df_minmax_scaled.to_csv('scale_v0.csv', index=False) # 저장
```
